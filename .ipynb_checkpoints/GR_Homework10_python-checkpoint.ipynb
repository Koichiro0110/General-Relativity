# Problem.3(iv)

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline

import scienceplots
plt.style.use('science')
%config InlineBackend.figure_format = 'retina'
```

```python
t_eval = np.linspace(0, np.pi, 10000)

# Differential equations
def equations(y, t):
    chi, v_chi = y
    # Equations of motion
    dchi_dt = v_chi
    dv_chidt = np.tan(chi) * (v_chi**2 - 1)
    
    return [dchi_dt, dv_chidt]
```

```python
y01 = [0.1*np.pi/2, 0.1]
y02 = [0.5*np.pi/2, 0.1]
y03 = [0.99*np.pi/2, 0.1]

y04 = [0.1*np.pi/2, 0.5]
y05 = [0.5*np.pi/2, 0.5]
y06 = [0.99*np.pi/2, 0.5]

y07 = [0.1*np.pi/2, 0.99]
y08 = [0.5*np.pi/2, 0.99]
y09 = [0.99*np.pi/2, 0.99]

sol1 = odeint(equations, y01, t_eval)
sol2 = odeint(equations, y02, t_eval)
sol3 = odeint(equations, y03, t_eval)
sol4 = odeint(equations, y04, t_eval)
sol5 = odeint(equations, y05, t_eval)
sol6 = odeint(equations, y06, t_eval)
sol7 = odeint(equations, y07, t_eval)
sol8 = odeint(equations, y08, t_eval)
sol9 = odeint(equations, y09, t_eval)
```

```python
sol1_chi  = sol1[:, 0]
sol2_chi  = sol2[:, 0]
sol3_chi  = sol3[:, 0]
sol4_chi  = sol4[:, 0]
sol5_chi  = sol5[:, 0]
sol6_chi  = sol6[:, 0]
sol7_chi  = sol7[:, 0]
sol8_chi  = sol8[:, 0]
sol9_chi  = sol9[:, 0]

plt.figure(figsize=(7, 5))

# Plot t vs τ
plt.axhline(y=np.pi/2, color='b', linestyle='--', label=r'$\chi = \frac{\pi}{2}$')
plt.plot(t_eval, sol1_chi, label=r'$\chi_0 = {:.2f}, v^\chi_0 = {:.2f}$'.format(y01[0], y01[1]))
plt.plot(t_eval, sol2_chi, label=r'$\chi_0 = {:.2f}, v^\chi_0 = {:.2f}$'.format(y02[0], y02[1]))
plt.plot(t_eval, sol3_chi, label=r'$\chi_0 = {:.2f}, v^\chi_0 = {:.2f}$'.format(y03[0], y03[1]))
plt.plot(t_eval, sol4_chi, label=r'$\chi_0 = {:.2f}, v^\chi_0 = {:.2f}$'.format(y04[0], y04[1]))
plt.plot(t_eval, sol5_chi, label=r'$\chi_0 = {:.2f}, v^\chi_0 = {:.2f}$'.format(y05[0], y05[1]))
plt.plot(t_eval, sol6_chi, label=r'$\chi_0 = {:.2f}, v^\chi_0 = {:.2f}$'.format(y06[0], y06[1]))
plt.plot(t_eval, sol7_chi, label=r'$\chi_0 = {:.2f}, v^\chi_0 = {:.2f}$'.format(y07[0], y07[1]))
plt.plot(t_eval, sol8_chi, label=r'$\chi_0 = {:.2f}, v^\chi_0 = {:.2f}$'.format(y08[0], y08[1]))
plt.plot(t_eval, sol9_chi, label=r'$\chi_0 = {:.2f}, v^\chi_0 = {:.2f}$'.format(y09[0], y09[1]))
plt.xlabel(r'$t$')
plt.ylabel(r'$\chi$')
plt.title(r'$\chi(t)$')
plt.legend(fontsize = 9)
```
