```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline

import scienceplots
plt.style.use('science')
%config InlineBackend.figure_format = 'retina'
```

```python
#tau_eval = np.linspace(0, 6*np.pi, 10000)
tau_eval = np.linspace(0, 20, 10000)

# Differential equations
def equations(y, tau):
    v, r, vp, rp  = y
    # Equations of motion
    dv_dtau = vp
    dr_dtau = rp
    dvp_dtau = - vp**2/r**2
    drp_dtau = - vp**2/r**2*(1 - 2/r) + 2*vp*rp / r**2
    
    return [dv_dtau, dr_dtau, dvp_dtau, drp_dtau]

y0 = [0, 6, 1, 0]
sol = odeint(equations, y0, tau_eval)
```

```python
sol_v  = sol[:, 0]
sol_r  = sol[:, 1]
sol_vp = sol[:, 2]
sol_rp = sol[:, 3]
# Plotting results
plt.figure(figsize=(7,5))

# t vs τ
plt.plot(tau_eval, sol_r, label=r'$r(\tau)$', color='red')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$r(\tau)$')
plt.title(r'$r(\tau)$')
```
